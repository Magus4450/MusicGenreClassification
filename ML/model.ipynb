````python
import numpy as np
import pandas as pd
````

````python
data = pd.read_csv("../data.csv")
data.head()
````

````python
labels = data['Genre'].unique()
labels
````

````python
data.describe()
````

````python
data.drop(columns=["Unnamed: 0", "Name"], inplace=True)
````

````python
# Replacing infinity values with NaN
data.replace([np.inf, -np.inf], np.nan, inplace=True)

````

````python
data.info()
````

### Filling NaN

````python
data["BER_Mean"].fillna(data["BER_Mean"].mean(), inplace=True)
data["BER_Std"].fillna(data["BER_Std"].mean(), inplace=True)
````

````python
print(f"Total rows having NaN values: {sum(data.isna().sum())}")
````

### Label Encoding

````python
# Label encoding the output
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Genre'] = le.fit_transform(data['Genre'])
````

````python
data['Genre'].value_counts()
````

````python
X = data.drop("Genre", axis=1)
y = data["Genre"]
````

### Oversampling

Since the label is not uniformly distributed, it may affect the perfomance of the model by giving less data for the model to generalize.

Oversampling fixes this by randomly distributing exampels from smaller class and adding them to the training dataset.

````python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state = 0)

X_oversampled, y_oversampled = ros.fit_resample(X, y)
````

````python
from collections import Counter
print(sorted(Counter(y_oversampled).items()))
````

### Normalizing
Normlizing helps models to converge faster

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_X = pd.DataFrame(scaler.fit_transform(X_oversampled), columns=X_oversampled.columns)
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_X, y_oversampled, test_size=0.2)
````

````python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
````

````python
from sklearn.metrics import classification_report
print("Training","-"*20)
print(classification_report(y_train, rfc.predict(X_train)))
print("Testing","-"*20)
print(classification_report(y_test, rfc.predict(X_test)))
````

With no hyperparameter tuning, the model seems to overfit.

````python
importances = rfc.feature_importances_
std = np.std([
    tree.feature_importances_ for tree in rfc.estimators_], axis=0)
````

````python
import matplotlib.pyplot as plt

forest_importances = pd.Series(importances, index=X_oversampled.columns)
forest_importances.sort_values(inplace=True, ascending=False)

fig, ax = plt.subplots()
forest_importances.plot.bar(yerr=std, ax=ax)
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
````

RMS_Std seems to be the most prominent feature with other features not far behind it.
